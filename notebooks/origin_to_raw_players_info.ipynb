```
from nba_api.stats.static import players

active_players = players.get_active_players()

active_players = spark.createDataFrame(active_players)
```

##Guardamos el dataframe como csv en el datalake

```
players_info.write.csv(
    path = f'abfss://{container}@{datalake}.dfs.core.windows.net/integrador/raw_data/players_info.csv',
    header = True,
    mode = 'overwrite'
    )

players_salaries.write.csv(
    path = f'abfss://{container}@{datalake}.dfs.core.windows.net/integrador/raw_data/players_salaries.csv',
    header = True,
    mode = 'overwrite'
    )
active_players.write.csv(
    path = f'abfss://{container}@{datalake}.dfs.core.windows.net/integrador/raw_data/active_players.csv',
    header = True,
    mode = 'overwrite'
    )
```
